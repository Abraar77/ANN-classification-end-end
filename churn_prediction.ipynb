```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import numpy as np
import pandas as pd

import pickle as pkl

```

```python
with open("label_enoder_gender.pkl", "rb") as f:
    label_encoder = pkl.load(f)
with open("one_hot_encoder_geo.pkl", "rb") as f:
    geo_encoder = pkl.load(f)
with open("scaler.pkl", "rb") as f:
    scaler = pkl.load(f)

model =load_model("model.h5")
```

```python
# Example input data
input_data = {
     'CreditScore': 600,
     'Geography': 'France',
     'Gender': 'Male', 'Age': 40,
     'Tenure': 3, 'Balance': 60000,
     'NumOfProducts': 2,
     'HasCrCard': 1,
     'IsActiveMember': 1,
     'EstimatedSalary': 50000
     }
```

```python
gen_enconded = geo_encoder.transform([[input_data["Geography"]]])
gen_enconded
```

```python
geo_encoder.categories_[0]
```

```python
geo_endoed_df= pd.DataFrame(gen_enconded, columns=geo_encoder.categories_[0])
geo_endoed_df
```

```python
input_data = pd.concat([pd.DataFrame([input_data]).drop("Geography", axis=1), geo_endoed_df], axis=1)
input_data
```

```python
input_data["Gender"]= label_encoder.transform(input_data["Gender"])
input_data
```

```python
input_data_scaled= scaler.transform(input_data)
input_data_scaled
```

```python
prediction = model.predict(input_data_scaled)
```

```python
prediction
```

```python
prediction_prob = prediction[0][0]
```

```python
prediction_prob
```

```python
if prediction_prob>0.5:
  print("coustomer is likely to leave")
else:
  print("coustomer is not likely to leave")
```
